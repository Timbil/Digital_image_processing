```python
import numpy as np
from matplotlib import pyplot as plt

def draw_polygon(ax, n):
    
    print("Введите координаты перемещения:")
    
    a=float(input("a="))
    b=float(input("b="))

    x = np.random.randint(0+a,50+a,n)
    y = np.random.randint(0+b,50+b,n)

  
    center_point = [np.sum(x)/n, np.sum(y)/n]

    angles = np.arctan2(x-center_point[0],y-center_point[1])

   
    sort_tups = sorted([(i,j,k) for i,j,k in zip(x,y,angles)], key = lambda t: t[2])

    
    if len(sort_tups) != len(set(sort_tups)):
        raise Exception('two equal coordinates -- exiting')

    x,y,angles = zip(*sort_tups)
    x = list(x)
    y = list(y)

    
    x.append(x[0])
    y.append(y[0])

    ax.plot(x,y, label = '{}'.format(n))

if __name__ == '__main__':

    fig,ax = plt.subplots()
    
    #n=int(input("Введите кол-во вершин: "))
    
    n=np.random.randint(3,50,1)
    
    
    print("Кол-во вершин", n)

    #for n in range(3,n):
    draw_polygon(ax,n)

    ax.legend()
    plt.show()
```
