```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
%matplotlib inline
```

```python
# создаем картинку c размером экрана 1920 х 1080 , фоном-белым, сохраняем 

img = Image.new('RGB', (1920, 1080), 'white')
img.save('test1.jpg')
```

```python
# функция которая , открывает картинку и закрашивает в нем пиксель[x][y] в черный цвет

def pixel_paint(x,y, img):
    img.putpixel((x, y), (253, 0, 0))
    img=img.save('test1.jpg')
```

```python

def draw_line(x1, y1, x2, y2, img):
         
    dx = x2 - x1
    dy = y2 - y1
        
    sign_x = 1 if dx>0 else -1 if dx<0 else 0
    sign_y = 1 if dy>0 else -1 if dy<0 else 0
        
    if dx < 0: dx = -dx
    if dy < 0: dy = -dy
        
    if dx > dy:
        pdx, pdy = sign_x, 0
        es, el = dy, dx
    
    else:
        pdx, pdy = 0, sign_y
        es, el = dx, dy
        
    x, y = x1, y1
        
    error, t = el/2, 0        
    
    
   # plt.figure(figsize=(100, 500))    
   # plt.scatter(x,y)
    
    #setPixel(x, y)
    
    pixel_paint(x,y, img)
        
    while t < el:
        error -= es
        if error < 0:
            error += el
            x += sign_x
            y += sign_y
        else:
            x += pdx
            y += pdy
        t += 1
       
        #plt.scatter(x,y)
        
        pixel_paint(x,y, img)
        


```

```python
# теперь сгенирируем 100 таких линий , 
# для этого сгенирируем лист содержийщий псевдослучайные числа( координаты линий)
from random import randint

x_1 = [randint(0,1600) for x in range(100)]

y_1 = [randint(0,900) for x in range(100)]

x_2 = [randint(0,1600) for x in range(100)]

y_2 = [randint(0,900) for x in range(100)]


```

```python
# вызовем нашу функцию 100 раз--> 100 таких линии

for i in range(0,10):
    draw_line(x_1[i],y_1[i],x_2[i],y_2[i] , img)
    
```
