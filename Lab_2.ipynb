```python
# Список для хранения кадров.
frames = []
 
for i in range(0, 24):
    # Открываем изображение каждого кадра.
    
    # поварачиваем  изображение на 15 градусов и добавляем его в список , примення билинейное интерполяцию
    im_rotate = img_bil.rotate(i*15, expand=True, resample=PIL.Image.BICUBIC)
    
    # Добавляем кадр в список с кадрами.
    frames.append(im_rotate)
 

# Берем первый кадр и в него добавляем оставшееся кадры.
frames[0].save(
    'homer.gif',
    save_all=True,
    append_images=frames[1:],  # Срез который игнорирует первый кадр.
    optimize=True,
    duration=100,
    loop=0
)
```
